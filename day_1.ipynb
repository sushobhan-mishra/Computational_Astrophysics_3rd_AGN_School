# If-Elif-else Statements or Control Statement

```python
temp = int(input('Enter temp:)'))

if temp >= 5000:
       print('hot star')
elif temp > 3000 and temp < 5000:
       print('star')
else:
       print('cold star')              
```

# loops - for loop / while loop

```python
# for loops
for i in range(10):
       print(i) # i is basically the elements in the range 0-9 but excluding 10
```

```python
for x in name:
       print(x)
```

```python
# while loop - conditional loop

count = 1
while count <= 5:
       print(count)
       count += 2
```

```python
for i in range(10):
       if i == 5:
              break
       print(i)
```

```python
for i in range(10):
       if i == 5:
              continue
       print(i)
```
