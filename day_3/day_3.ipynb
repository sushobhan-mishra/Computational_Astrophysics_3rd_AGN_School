### Constants and Units

![Redshift](redshift.png)


```python
import astropy.units as u
import astropy.constants as const

def redshift(lam_obs,lam_th):
       lam_obs = lam_obs * u.nm
       lam_th = lam_th * u.nm

       #feq_obs = lam_obs.to(u.Hz, equivalence=u.spectral())
       z = (lam_obs - lam_th) / lam_th
       
       return z
```

```python
z1 = redshift(1500,1000)
z1
```

```python
vel = z1 * const.c
vel
```

```python
#v = h0d
dist = vel/70000
dist
```

relation betwwen apparent magnitude of star and absolute magnitude of star and distance of the star

```python
import astropy.units as u

def distance(m_app,m_abs):
       dm = m_app - m_abs
       d_pc = (10* 10**(dm/5))
       d_ly = (d_pc*u.parsec).to(u.lyr)

       return d_ly

distance(23,14)
```
