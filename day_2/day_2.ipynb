# PANDAS

```python
import csv

data = [
       ['Galaxy','Redshift'],
       ['M31','0.000282'],
       ['M33','0.000921'],
       ['M33','0.000921'],
       ['NGC7331',''],
       ['',''],
       ['NGC5907','0.00224'],
       ['NGC891','0.0025']
]

with open('galaxies.csv', mode = 'w', newline = '') as file:
       wr = csv.writer(file)
       wr.writerows(data)
       # wr is variable name which is defined to save the file as csv
       # wr.writerows is writing the rows from (data) in that csv file
```

```python
import pandas as pd
df = pd.read_csv('galaxies.csv')
df.head() # will print first five values
```

```python
df.isna() #nan is true
```

```python
df.dropna() # nan is dropped
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.rename(columns={'Redshift':'New_Redshift'},inplace=True)
```

```python
df
```

```python
df.fillna({'New_Redshift': np.mean(df['New_Redshift'])}, inplace=True)
```

```python
df.to_csv('clean_galaxy_data.csv', index=False)
```

# Matplotlib - For Ploting and Visualization

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

y = np.sin(x)

plt.figure(figsize=(6,6)) # (width,height)
plt.grid()
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Our First Plot - Sin(x)')
plt.plot(x,y)
plt.savefig('sinx-firstplt.pdf',dpi=1000)
plt.show()
```

```python
x = np.linspace(1,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,'-g',label = 'sinx')
plt.plot(x,y2,':b',label = 'cosx')
plt.grid()
plt.axis('equal')
plt.legend()
plt.show()
```

```python
data = np.random.normal(0,1,1000)
plt.hist(data,bins=30,color='lightblue')  #bins='auto'
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(6,6)) # (width,height)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterd Plot')
plt.scatter(x,y)
plt.show()
```

### Subplots

```python
# using plt.subplot
x = np.linspace(1,10,100)
plt.figure(figsize=(7,4))
plt.subplot(2,1,1) # (rows,colums,panel number)
plt.plot(x , np.sin(x),'-g', label='sinx')
plt.legend()

plt.subplot(2,1,2) # panel number changes
plt.plot(x , np.cos(x),':b',label='cosx')
plt.legend()
plt.show()
```

```python
# axs method
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axs = plt.subplots(2,2) # 2x2 matrix
fig.set_size_inches(7,4)

#axs[i,j]
axs[0,0].plot(x,y1,color='pink')
axs[0,1].plot(x,y2,color='black')
axs[1,0].plot(x,y3,color='green')
axs[1,1].plot(x,y4,color='blue')
plt.show()
```
